```python
import os
import pandas as pd

# Define the mappings for column values
label_mapping = {'COL': 0, 'CL': 1, 'COH': 2, 'NROI': -1}

# Define the folder containing the CSV files
folder_path = "C:/Users/kharep/Downloads/July_Pressure_data__rnn"
testing_folder_path =  "C:/Users/kharep/Downloads/July_Pressure_data_testing_rnn"
# Iterate over each CSV file in the folder
```

```python
def convert(folder_path):
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            
            # Read the CSV file
            df = pd.read_csv(file_path)
            
            # Preserve the column headers
            column_headers = list(df.columns)
            
            # Modify the last column based on the mappings, accounting for existing -1 values
            df.iloc[:, -1] = df.iloc[:, -1].map(lambda x: label_mapping[x] if x in label_mapping else -1)
            df.iloc[:, -2] = df.iloc[:, -2].map(lambda x: label_mapping[x] if x in label_mapping else -1)
            df.iloc[:, -3] = df.iloc[:, -3].map(lambda x: label_mapping[x] if x in label_mapping else -1)
            # Write the modified DataFrame back to the CSV file, preserving the column headers
            df.to_csv(file_path, index=False, header=column_headers)
```

```python
convert(folder_path)
```

```python
convert(testing_folder_path)
```
